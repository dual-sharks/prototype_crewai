```python
from crewai import Agent, Task, Crew, LLM
from crewai_tools import SerperDevTool
from litellm import completion
from env import (AWS_ACCESS_KEY_ID, AWS_SECRET_ACCESS_KEY, OPENAI_API_KEY)
```

```python
import os
os.environ["AWS_ACCESS_KEY_ID"] = AWS_ACCESS_KEY_ID
os.environ["AWS_SECRET_ACCESS_KEY"] = AWS_SECRET_ACCESS_KEY
os.environ["AWS_REGION_NAME"] = "us-east-2"
os.environ["OPENAI_API_KEY"] = ""
#os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
```

```python
from langchain_aws import BedrockLLM

custom_llm = BedrockLLM(
    credentials_profile_name="sharks",
    provider="bedrock",
    model_id="amazon.titan-tg1-large",
    model_kwargs={"temperature": 1},
    streaming=False,
)

custom_llm.invoke(input="What is the recipe of mayonnaise?")
```

```python
social_media_analyst = Agent(
 role='Social Media Analyst',
 goal='Review companies from a social media perspective in order to understand public sentiment and social media trends. You specifically want to understand how the company is seen on Twitter, Facebook, and other social media platforms.',
 backstory="You work at a leading social media marketing company.",
 verbose=True,
 allow_delegation=False,
 custom_llm=custom_llm,
#tool=[SerperDevTool()] # Tool for online searching
)
```

```python
researcher = Agent(
 role='Senior FINRA Analyst',
 goal='Uncover new information about companies from a financial perspective to be able to make investment decisions for stakeholders. You want to identify good investment opportunities and recommend potential strategies to benefit from the companies current position.',
 backstory="You work at a leading investment think tankâ€¦",
 verbose=True,
 allow_delegation=False,
 custom_llm=custom_llm,
#tool=[SerperDevTool()] # Tool for online searching
)
# Define the Task
task1 = Task(
 description="Conduct a comprehensive analysis of the following topic: {topic}",
 expected_output="Full analysis report in bullet points",
 agent=researcher
)
task2 = Task(
 description="Conduct a comprehensive social media and public sentiment analysis of the following topic: {topic}",
 expected_output="Full market report in bullet points",
 agent=social_media_analyst
)
crew = Crew(
 agents=[researcher, social_media_analyst],
 tasks=[task1, task2],
 verbose=True,
)
```

```python
result = crew.kickoff(inputs={"topic":"Recent News concerning the company NVIDIA"})
```
